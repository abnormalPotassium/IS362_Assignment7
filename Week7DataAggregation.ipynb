## Normalizing the Dataset by Name

We can rescale the data grouping by user and then using normalization in a range of [0,1]. This is helpful as different users will have different rating scales and gives us more meaningful data when we have the ratings adjusted so how all the users rate things are normalized.

```python
normalized_values = df_movies.groupby(['Name'])['Rating'].transform(lambda x: ((x-x.min())/(x.max()-x.min())))
df_movies_normalized = df_movies
df_movies_normalized.Rating = normalized_values
df_movies_normalized
```

### Normalized Average Movie Scores

```python
df_movies_normalized.groupby(['Movie']).mean()
```

### Normalized Average Name Scores

```python
df_movies_normalized.groupby(['Name']).mean()
```

## Why Used Normalized Ratings

As I explained before it makes sense to normalize ratings by users. Each user will have their own review scale that may skew towards rating movies highly or lowly. Using normalization makes it as if there is a standarized scale that the users are using because the lowest score of each user becomes 0 and the highest score becomes 1. This gets rid of skew towards scoring higher or lower which is advantageous in getting rid of some bias. However, this can be disadvantageous by bringing down the score of good movies and bringing up the score of bad movies. Many people only review things when they have strong emotions towards it and someone who only reviews movies they truly feel are good would have their scores not reflect how they felt. While the opposite problem might happen for someone who only reviews movies if they really don't like them.
